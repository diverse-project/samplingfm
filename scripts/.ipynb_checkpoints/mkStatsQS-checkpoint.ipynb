```python
import re
file = open("../out/FrequencyDiagrams/FM-3.6.1-refined_qs_log.txt", "r")
fts = pd.DataFrame(columns=['ft', 'obs', 'th', 'freqDev', 'rfreqDev'])

endOfFeatureAnalysis = False
idft = 0

for line in file:
#    m = re.search('^[0-9]+\s[a-zA-Z]+', line)
    m = re.search('(?<=[0-9] )(.*)', line)
#    ft = 
    if (not m is None):
        ft = m.group(0)
    
    m2 = re.search("(?<= obs :)(.*)", line) 
    if (not m2 is None):
        obs = float(m2.group(0))
    
    m3 = re.search("(?<= th  :)(.*)", line) 
    if (not m3 is None):
        th = float(m3.group(0))
        endOfFeatureAnalysis = True
        
    if (endOfFeatureAnalysis):
        endOfFeatureAnalysis = False        
        freqDev = abs(th - obs) * 100 / th
        rfreqDev = (obs - th) * 100 / th # without absolute
        fts.loc[idft] = (ft, obs, th, freqDev, rfreqDev)
        idft = idft + 1 
    
        
fts.sort_values(by="freqDev", ascending=False)

```

```python
# Hibernate2ndLvlCache
# ft == 'UaaServer' is out (not authentification!)
ftsFinal = fts.query("ft == 'MongoDB' | ft == 'MariaDB' | ft == 'Uaa' | ft == 'SocialLogin' | ft == 'Gradle' | ft == 'MariaDB' | ft == 'gradle' | ft == 'EhCache' | ft == 'Monolithic' | ft == 'Hibernate2ndLvlCache' | ft == 'ElasticSearch' | ft == 'Docker' | ft == 'Docker'")
ftsFinal['obs'] = round(ftsFinal['obs'], 3) 
ftsFinal['th'] = round(ftsFinal['th'], 3)
ftsFinal['freqDev'] = round(ftsFinal['freqDev'])
ftsFinal['rfreqDev'] = round(ftsFinal['rfreqDev'])

ftsFinal = ftsFinal.sort_values(by="freqDev", ascending=False)
ftsFinal.rename(columns={'ft' : 'feature', 'th' : 'fth', 'obs' : 'fobs', 'freqDev' : 'dev', 'rfreqDev' : 'rdev'}, inplace=True)
```

```python
with open('jhipster-fts.tex', 'w') as tf:
     tf.write(ftsFinal.to_latex(index=False))
```
